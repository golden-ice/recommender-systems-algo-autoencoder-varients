```python
import os
import json
import pandas as pd
import numpy as np
```

```python
# generate ml-explicit-10kplus
```

```python
file_dir = '/content/gdrive/My Drive/msci720_prj/dataset/train-dataset/ml-explicit'
file = 'ratings.csv'
file_path = os.path.join(file_dir, file)
```

```python
columns = ["userId", "movieId", "rating", "timestamp"]
ratings_train = pd.read_csv(file_path, sep=',')
```

```python
ratings_train
```

```python
file_dir = '/content/gdrive/My Drive/msci720_prj/'
file = 'userIds.json'
file_path = os.path.join(file_dir, file)
```

```python
with open(file_path, "r") as json_file:
    userIds_pred = json.load(json_file)
```

```python
user_train_fromPred = pd.DataFrame({'userId': userIds_pred})
```

```python
training_set_userPred = pd.merge(ratings_train, user_train_fromPred, on=["userId"], how="inner")
```

```python
train_userIds = ratings_train[['userId']].drop_duplicates()
```

```python
train_userIds
```

```python
train_userId_filtered = train_userIds[~train_userIds['userId'].isin(user_train_fromPred['userId'])]
```

```python
train_userId_filtered
```

```python
train_random_userIds = train_userId_filtered['userId'].sample(n=10000, random_state=42)
```

```python
train_random_userIds
```

```python
random_userIds_sorted = train_random_userIds.sort_values()
```

```python
sorted_userIds_list = random_userIds_sorted.tolist()
```

```python
with open('/content/gdrive/My Drive/msci720_prj/10k_userIds_train.json', 'w') as f:
    json.dump(sorted_userIds_list, f)
```

```python
file_dir = '/content/gdrive/My Drive/msci720_prj/'
file = 'userIds.json'
file_path = os.path.join(file_dir, file)
with open(file_path, "r") as json_file:
    userIds_pred = json.load(json_file)
```

```python
file_dir = '/content/gdrive/My Drive/msci720_prj/'
file = '10k_userIds_train.json'
file_path = os.path.join(file_dir, file)
with open(file_path, "r") as json_file:
    userIds_10k = json.load(json_file)
```

```python
merged_userIds = userIds_pred + userIds_10k
# Create a DataFrame from the merged list
df_merged = pd.DataFrame({'userId': merged_userIds})
df_merged
```

```python
training_set_user11k = pd.merge(ratings_train, df_merged, on=["userId"], how="inner")
```

```python
training_set_user11k
```

```python
save_dir = '/content/gdrive/My Drive/msci720_prj/dataset/train-dataset/ml-explicit'
training_set_selected_name = 'training_set_user11k.csv'
training_set_user11k_savepath = os.path.join(save_dir, training_set_selected_name)

training_set_user11k.to_csv(training_set_user11k_savepath, sep=',', header=True, index=False)
```
